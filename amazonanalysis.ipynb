```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('Amazon Sales data.csv')
data=pd.DataFrame(data)
print(data)
```

```python
data = data.astype({'Ship Date': 'datetime64[ns]','Order Date':'datetime64[ns]'})
```

```python
data.dtypes
```

```python
data.isnull()
```

```python
Highest_Total_Revenue= data.groupby(data['Region'])['Total Revenue'].sum()
Highest_Total_Revenue.idxmax()
```

```python
plt.figure(figsize=(20,8))
sns.lineplot(Highest_Total_Revenue,linestyle= '--' ,linewidth= 4 , marker= 'o', markersize= 10, 
             markerfacecolor='red', markeredgecolor='black')
plt.title('To Show Highest Total Sales Revenue by Region', fontsize= 20)
plt.xlabel('Region', fontsize=10)
plt.ylabel('Total Revenue', fontsize=10)
plt.show()
```

```python
avg_unit_price= data.groupby(data['Item Type'])['Unit Price'].mean()
avg_unit_cost= data.groupby(data['Item Type'])['Unit Cost'].mean()
average= pd.DataFrame({'avg_unit_price':avg_unit_price,
                      'avg_unit_cost':avg_unit_cost})
print(average)
```

```python
Highest_total_profit=data.groupby(data['Country'])['Total Profit'].max()
Highest_total_profit
```

```python
Highest_total_profit=data.groupby(data['Country'])['Total Profit'].sum()
highest_country_profit=Highest_total_profit.idxmax()
highest_country_profit
```

```python
plt.figure(figsize=(20,10))
sns.barplot(Highest_total_profit)
plt.title('Highest Total Profit')
plt.xticks(rotation=90)
```

```python
data['Sales Channel'].value_counts().plot(kind='pie',autopct='%1.1f%%',title='Channel Sales')
```

```python
item_type=data.groupby(data['Item Type'])['Total Revenue'].sum()
highest_cost_item=item_type.idxmax()
lowest_cost_item=item_type.idxmin()
print("{'Highest Sales Revenue By Item Type':", highest_cost_item, "\n'Lowest Sales Revenue By Item Type':", lowest_cost_item, "}")
```

```python
diff_region_order_priority= data.groupby(data['Region'])['Order Priority'].value_counts()
print(diff_region_order_priority)

```

```python
plt.figure(figsize=(20,10))
sns.countplot(x='Region',hue='Order Priority',data=data)
plt.ylabel('order priority count')
plt.show()

```

```python
unit_sold=data.groupby(data['Country'])['Units Sold'].sum()
unit_sold
```

```python
plt.figure(figsize=(20,10))
sns.barplot(data=unit_sold)
plt.xticks(rotation='vertical')
```

```python
country_sales=data.groupby(data['Country'])['Total Revenue'].sum()
print(country_sales)
```

```python
plt.figure(figsize=(20,10))
sns.barplot(data=country_sales)
plt.title('Country Sales')
plt.xticks(rotation='vertical')
```
